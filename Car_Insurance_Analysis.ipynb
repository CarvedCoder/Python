```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
df = pd.read_csv('automobile_dataset.csv')
df.info()
```

```python
df.describe()
```

```python
numeric_cols = len(df.columns[4:])
numeric_rows = math.ceil(numeric_cols/3)

# create subplots with a grid of numeric_rows * 3 

fig , axes = plt.subplots(numeric_rows,3,figsize=(15,5*numeric_rows))
axes = axes.flatten()

#loop through the columns and plot histograms

for i,col in enumerate(df.columns[4:]):
    axes[i].hist(df[col],bins=20,color= "blue",alpha=0.7)
    axes[i].set_title(f"Histogram of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Frequency")

for j in range(i+1,len(axes)):
    fig.delaxes(axes[j])
    
plt.tight_layout()
plt.show()
```
